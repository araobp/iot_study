Measurement date: August 14, 2018

Location: second floor of my house

Heat the following materials under strong sunshine:

- White metal plate w/ or w/o alminium heat sink

```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.signal as sg

matplotlib.rcParams['figure.figsize'] = [16, 6]
FILE1 = 'temperature3.csv'
FILE2 = 'temperature4.csv'
```

```python
df = pd.read_csv(FILE1)
df.head()
```

```python
df2 = pd.read_csv(FILE2)
df['black metal heat sink'] = df2['black metal']
df['white metal heat sink'] = df2['white metal']
df.head()
```

```python
ax = df[['black metal', 'white metal', 'black metal heat sink', 'white metal heat sink']].plot(x=df['elapsed time'], title='Temperature')
ax.set_xlabel('Elaplsed time in sec')
ax.set_ylabel('Temperature in degrees Celsius')
```
