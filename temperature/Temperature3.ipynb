```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.signal as sg

matplotlib.rcParams['figure.figsize'] = [16, 6]
FILE1 = 'temperature5.csv'
FILE2 = 'temperature6.csv'
FILE3 = 'temperature7.csv'
FILE4 = 'temperature8.csv'
```

```python
df = pd.read_csv(FILE1)
df.head()
```

```python
ax = df[['alminium bar', 'iron bar', 'copper bar']].plot(x=df['elapsed time'], title='Temperature at 10cm away from the hot water')
ax.set_xlabel('Elaplsed time in sec')
ax.set_ylabel('Temperature in degrees Celsius')
```
