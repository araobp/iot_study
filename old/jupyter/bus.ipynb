# Actual measurement performed on a bus in Japan

- Three-axis acceleration sensor IC: Kionix KXR94-2050 (http://akizukidenshi.com/catalog/g/gM-05153/)
- Measurement period: around 500 msec

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
TITLE = "Three-axis accelerometer"
```

```python
df = pd.read_csv(r"C:\Users\shiny\MPLABXProjects\blocks\jupyter\accelerometer.csv")
```

### Mesasurement summary

```python
df.describe()
```

### Data range: 0 ~ 800

```python
p = df[['acceleration.x', 'acceleration.y', 'acceleration.z']][0:800].plot(title=TITLE, figsize=(17,6), grid=True)
p.set_xlabel('Time count (every 500msec)')
p.set_ylabel('Force (G)')
```
