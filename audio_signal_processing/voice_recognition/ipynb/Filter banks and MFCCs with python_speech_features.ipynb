# Filter banks and MFCCs with python_speech_features

reference: https://python-speech-features.readthedocs.io/en/latest/

```python
from python_speech_features import mfcc
from python_speech_features import logfbank
import scipy.io.wavfile as wav
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

NFILT=40
NUMCEP=20
FILE='aiueo.wav'

rate, sig = wav.read(FILE)
mfcc_feat = mfcc(sig, rate, numcep=NUMCEP, nfilt=NFILT, winlen=0.0115)
fbank_feat = logfbank(sig, rate, nfilt=NFILT, winlen=0.0115)
mfcc_feat.shape, fbank_feat.shape
```

```python
f = np.arange(0, fbank_feat.shape[1])
t = np.linspace(0, len(sig)/rate,fbank_feat.shape[0])

plt.figure(figsize=[12,3])
plt.pcolormesh(t, f[:25], fbank_feat.T[:25], cmap='ocean')
plt.xlabel('Time [sec]')
plt.ylabel('Filter bank features')
plt.title("Features")
```

```python
c = np.arange(0, NUMCEP+1)
plt.figure(figsize=[12,3])
plt.pcolormesh(t, c[1:13], mfcc_feat.T[1:13], cmap='hot')
plt.xlabel('Time [sec]')
plt.ylabel('MFCC coefficients')
plt.title("MFCCs")
```
